```python
df=df.drop(columns=["ROI"])
```

```python
df=df.rename(columns={'Avg_Likes':'Likes','Avg_Comments':'Comments','Avg_Shares':'Shares'})
```

```python
df['Campaign_Cost']=df['Campaign_Cost'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
df['Revenue']=df['Revenue'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
```

```python
df['Campaign_Cost'] = pd.to_numeric(df['Campaign_Cost'], errors='coerce')
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')
```

```python
df
```

```python
# Replace all NaNs in a Likes and Comments Columns with 0
df['Likes'] = df['Likes'].fillna(0)
```

```python
df['Comments'] = df['Comments'].fillna(0)
```

```python
# Filling missing values in Followers column with Median of their specific Influencer_Tier
df['Followers'] = df.groupby('Influencer_Tier')['Followers'].transform(lambda x: x.fillna(x.median()))
```

```python
# Filling missing values in Campaign_Cost column with Median of their specific Influencer_Tier
df['Campaign_Cost'] = df.groupby('Influencer_Tier')['Campaign_Cost'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Strip hidden spaces from ID columns
df['Campaign_ID'] = df['Campaign_ID'].str.strip()
df['Influencer_ID'] = df['Influencer_ID'].str.strip()
```

```python
#Check and remove duplicate rows
print(f"Duplicates before: {df.duplicated().sum()}")
```

```python
df = df.drop_duplicates().reset_index(drop=True)
print(f"Duplicates after: {df.duplicated().sum()}")
```

```python
df
```

```python
# Convert count-based columns to Integers
# (Since you've handled NaNs, we can use int64)
int_cols = ['Followers', 'Likes', 'Comments', 'Shares', 'Revenue']
for col in int_cols:
    df[col] = df[col].astype('int64')
```

```python
# Convert repeating text labels to Categories
# This is a best practice for columns like Platform and Tier
cat_cols = ['Platform', 'Influencer_Tier', 'Campaign_Type']
for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
# Convert IDs to explicit Strings
df['Campaign_ID'] = df['Campaign_ID'].astype('string')
df['Influencer_ID'] = df['Influencer_ID'].astype('string')
```

```python
# Final verification
print(df.info())
```

```python
# Strip whitespace from all string columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Standardize Influencer_Name to Title Case
df['Influencer_Name'] = df['Influencer_Name'].str.title()
```

```python
# Check null counts
print(df.isnull().sum())
```
